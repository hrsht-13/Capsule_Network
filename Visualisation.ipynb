**Visualisation of the results obtained after training the data with both 2 and 3 routing iterations**



```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
num2=pd.read_csv("/content/CPcsvNum2 - Sheet1.csv")
```

```
num3=pd.read_csv("/content/CPcsvnum3 - Sheet1 (1).csv")
```

```
num3.columns
```

```
ax=sns.lineplot(x="epoch",y="loss",data=num2)
px=sns.lineplot(x="epoch",y="loss",data=num3)
plt.legend(["routing=2","routing=3"],loc="upper right")
plt.title("loss vs epochs for different routing iterations")
plt.xlabel('Epochs')
plt.ylabel('loss(encoder+0.0005*decoder)')
```

```
d1=sns.lineplot(x="epoch",y="encoder_acc",data=num3)
d2=sns.lineplot(x="epoch",y="val_encoder_acc",data=num3)
plt.legend(["Training","Validation"],loc="upper right")
plt.title("Accuracy vs epoch")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

```
num3["val_decoder_loss"]=num3["val_decoder_loss"]/0.0005
```

```
d1=sns.lineplot(x="epoch",y="encoder_loss",data=num3)
d2=sns.lineplot(x="epoch",y="val_encoder_loss",data=num3)
plt.legend(["Training","Validation"],loc="upper right")
plt.title("loss vs epoch")
plt.xlabel('Epochs')
plt.ylabel('loss')
```

```
d1=sns.lineplot(x="epoch",y="decoder_loss",data=num3)
d2=sns.lineplot(x="epoch",y="val_decoder_loss",data=num3)
plt.legend(["Training","Validation"],loc="upper right")
plt.title("loss vs epoch")
plt.xlabel('Epochs')
plt.ylabel('loss')
```

```
d1=sns.lineplot(x="epoch",y="encoder_acc",data=num3)
d2=sns.lineplot(x="epoch",y="val_encoder_acc",data=num3)
plt.legend(["Training","Validation"],loc="upper right")
plt.title("ENCODER")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
```

```
d1=sns.lineplot(x="epoch",y="val_encoder_loss",data=num3)
d2=sns.lineplot(x="epoch",y="val_decoder_loss",data=num3)
plt.legend(["encoder","decoder"],loc="upper right")
plt.title("Training loss")
plt.xlabel('Epochs')
plt.ylabel('loss')
```
